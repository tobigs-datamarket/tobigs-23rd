## 빈칸을 채워서 모델을 로드해주세요

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "skt/kogpt2-base-v2"
tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    bos_token="</s>",
    eos_token="</s>",
    unk_token="<unk>",
    pad_token="<pad>",
    mask_token="<mask>"
)
model = AutoModelForCausalLM.from_pretrained(model_name)
model
```

```python
from datasets import load_dataset

split_dict = {
    "train": "train[:8000]",
    "test": "train[8000:10000]",
    "unused": "train[10000:]",
}
dataset = load_dataset("heegyu/kowikitext", split=split_dict) # custom code y 입력해주세요
del dataset["unused"]
dataset
```

## 아래 train, test Dataset 에서 feautures 중 input_ids, attention_mask 각각의 역할을 서술해주세요

## 정답:
- input_ids: 입력 텍스트를 토큰화하여 숫자 형태로 변환한 값이다. 각 토큰은 모델의 어휘(Vocabulary)에 있는 고유한 숫자 ID로 매핑되며, 이 값은 모델이 텍스트를 이해하고 처리하는 데 사용된다. 예를 들어, 문장이 "지난해 7월,"이라면, 이 문장을 토큰화하여 각 단어를 숫자로 변환한 결과가 input_ids에 저장된다.

- attention_mask: 모델이 처리할 토큰과 무시할 토큰을 구분하는 이진 마스크이다. 1은 실제 택스트(입력) 토큰을 의미하고, 0은 패딩 토큰을 의미한다. 패딩 토큰은 모델 학습에 영향을 미치지 않도록 무시되며, attention_mask는 모델에게 패딩을 건너뛰고 실제 텍스트만 처리하라고 지시한다. 예를 들어, 긴 문장에서 부족한 길이를 채우기 위해 추가된 패딩 토큰이 있을 때, 이들을 무시하고 실제 의미 있는 텍스트만 학습하도록 돕는다.


```python
tokenized_dataset = dataset.map(
    lambda batch: tokenizer([f"{ti}\n{te}" for ti, te in zip(batch["title"], batch["text"])]),
    batched=True,
    num_proc=2,
    remove_columns=dataset["train"].column_names,
)
tokenized_dataset
```

```python
max_length = 512
def group_texts(batched_sample):
    sample = {k: v[0] for k, v in batched_sample.items()}

    if sample["input_ids"][-1] != tokenizer.eos_token_id:
        for k in sample.keys():
            sample[k].append(
                tokenizer.eos_token_id if k == "input_ids" else sample[k][-1]
            )

    result = {k: [v[i: i + max_length] for i in range(0, len(v), max_length)] for k, v in sample.items()}
    return result

grouped_dataset = tokenized_dataset.map(
    group_texts,
    batched=True,
    batch_size=1,
    num_proc=2,
)
print(len(grouped_dataset["train"][0]["input_ids"]))
print(grouped_dataset)
```

```python
from transformers import DataCollatorForLanguageModeling

collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
sample = collator([grouped_dataset["train"][i] for i in range(1)])
```

```python
import os
import tempfile
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
from langchain_community.document_loaders import TextLoader
from langchain.prompts import PromptTemplate
from google.colab import files
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline, BitsAndBytesConfig
```

```python
# 파일 업로드 함수
def upload_file():
    print("문서 파일을 업로드해주세요.")
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
    print(f"업로드 완료: {file_path}")
    return file_path
```

```python
# LLM 모델 로컬 구성
def get_local_llm():
    print("언어 모델 로딩 중... (처음 실행 시 모델 다운로드에 시간이 걸릴 수 있습니다)")
    model_id = "google/flan-t5-base"  # 더 작은 모델로 변경

    quantization_config = BitsAndBytesConfig(
        load_in_8bit=True,
        llm_int8_enable_fp32_cpu_offload=True
    )
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_id, device_map="auto" # 수정된 부분
)

    pipe = pipeline(
        "text2text-generation",
        model=model,
        tokenizer=tokenizer,
        max_length=512,
        temperature=0.5
    )

    llm = HuggingFacePipeline(pipeline=pipe)
    print("언어 모델 로딩 완료!")
    return llm
```

## 아래 코드를 참고하여 RAG 프로세스를 서술해주세요

## 정답 :
** RAG (Retrieval-Augmented Generation)
1. 문서 로드

- 주어진 파일 경로에서 문서를 로드한다. "TextLoader" 클래스를 사용하여 텍스트 파일을 로드하고, 이를 documents라는 변수에 저장한다.
- RAG 시스템에서 사용할 원본 문서를 불러오는 첫 번째 단계이다.

2. 문서 분할

- 로드된 문서를 더 작은 청크로 분할한다. RecursiveCharacterTextSplitter를 사용하여 문서를 일정 크기(chunk_size=1000)로 나누고, 각 청크가 겹치도록(chunk_overlap=100)한다.
- 긴 문서를 작고 처리 가능한 크기로 나누어 검색 및 모델 학습에 적합하도록 만든다.

3. 임베딩 생성 및 벡터 저장소 구축

- 문서 청크에 대한 임베딩을 생성한다. HuggingFaceEmbeddings를 사용하여 각 텍스트 청크를 벡터 형태로 변환하고, FAISS 라이브러리를 사용해 벡터 저장소를 구축한다.
- 텍스트를 벡터 공간으로 변환하여, 유사한 텍스트를 효율적으로 검색할 수 있게 한다.

4. 검색기 생성

- FAISS로 구축한 벡터 저장소에서 문서를 검색할 수 있는 검색기를 생성한다. k=3으로 설정하여, 검색 시 가장 유사한 상위 3개의 문서를 반환한다.
- 사용자가 입력한 질문과 관련된 문서를 검색하기 위해 검색기를 설정한다.

5. LLM 설정 (로컬 파이프라인 사용)
- 질문에 대한 답변 생성을 위한 언어 모델(LLM)을 설정한다. get_local_lim()을 사용하여 로컬에서 실행 가능한 LLM을 불러온다.
- 질문에 대한 답변을 생성하는 언어 모델을 설정하는 단계이다.

6. 프롬프트 템플릿 정의
- 질문과 관련된 정보를 기반으로 답변을 생성하기 위한 프롬프트 템플릿을 정의한다. {context}에는 검색된 문서가 들어가고, {question}에는 사용자가 입력한 질문이 들어간다.
- 모델에게 정보를 어떻게 전달할지에 대한 형식을 정의한다.

7. RAG 파이프라인 구축
- RAG 시스템을 구축한다. RetrievalQA를 사용하여 검색된 문서(retriever)와 언어 모델(LLM)을 연결하여, 최종적으로 질문에 대한 답변을 생성할 수 있는 파이프라인을 만든다. 이때 chain_type = 'stuff'는 검색된 모든 문서를 사용하여 답변을 생성하도록 설정한다.
- 전체적인 RAG 시스템을 구성하여, 질문에 대한 답변을 생성할 수 있도록 한다.

8. 대화형 인터페이스
- 사용자로부터 질문을 입력받고, RAG 시스템을 사용하여 해당 질문에 대한 답변을 생성한다. 사용자가 q를 입력할 때까지 반복한다.
- 사용자와의 인터페이스를 제공하여, 실시간으로 질문과 답변을 주고받을 수 있도록 한다.

9. 벡터 저장소 저장
- 벡터 저장소를 로컬에 저장한다. 임시 디렉토리를 사용하여 FAISS 인덱스를 저장하고, 다운로드할 수 있는 링클르 제공한다.
- 생성된 벡터 인덱스를 저장하고, 나중에 다시 사용할 수 있도록 한다.


--> 전체 프로세스는 질문에 대해 관련된 문서를 검색하고, 해당 문서를 바탕으로 답변을 생성하는 RAG 시스템의 핵심 흐름을 구현한다.

```python
# RAG 시스템 구현
def colab_rag_system(file_path):
    # 1. 문서 로드
    loader = TextLoader(file_path)
    documents = loader.load()

    print(f"문서 로딩 완료: {len(documents)} 개의 문서")

    # 2. 문서 분할
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=100,
        length_function=len
    )
    texts = text_splitter.split_documents(documents)

    print(f"문서 분할 완료: {len(texts)} 개의 청크")

    # 3. 임베딩 생성 및 벡터 저장소 구축
    print("임베딩 모델 로딩 중...")
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

    print("벡터 저장소 구축 중...")
    vectorstore = FAISS.from_documents(texts, embeddings)

    # 4. 검색기 생성
    retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

    # 5. LLM 설정 - 로컬 파이프라인 사용
    llm = get_local_llm()

    # 6. 프롬프트 템플릿 정의
    prompt_template = """
    다음 정보를 바탕으로 질문에 답해주세요:

    {context}

    질문: {question}
    답변:
    """

    PROMPT = PromptTemplate(
        template=prompt_template,
        input_variables=["context", "question"]
    )

    # 7. RAG 파이프라인 구축
    print("RAG 파이프라인 구축 중...")
    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever,
        return_source_documents=True,
        chain_type_kwargs={"prompt": PROMPT}
    )

    print("RAG 시스템 준비 완료!")

    # 8. 대화형 인터페이스
    while True:
        query = input("\n질문을 입력하세요 (종료하려면 'q' 입력): ")

        if query.lower() == 'q':
            break

        result = qa_chain({"query": query})

        print("\n답변:", result["result"])
        print("\n관련 문서 출처:")
        for i, doc in enumerate(result["source_documents"]):
            print(f"문서 {i+1}: {doc.page_content[:100]}...")

    # 9. 벡터 저장소 저장
    with tempfile.TemporaryDirectory() as temp_dir:
        index_path = os.path.join(temp_dir, "faiss_index")
        vectorstore.save_local(index_path)
        print(f"\n인덱스를 '{index_path}'에 저장했습니다.")

        # 인덱스 다운로드 제공
        print("벡터 인덱스를 다운로드하려면 아래 코드를 실행하세요:")
        print("files.download(f'{index_path}/index.faiss')")
        print("files.download(f'{index_path}/index.pkl')")
```
